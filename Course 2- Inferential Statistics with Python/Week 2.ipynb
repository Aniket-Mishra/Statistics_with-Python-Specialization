## Sampling Distribution of Sample Mean

If model for population of responces is approximately normal (or sample size is large enough), distribution of sample mean is approximately normal. ( All possible values of sample mean), so we are talking about unbiased estimation<br>
We need to have a better understanding of the standard error (the spread of the distribution) of the sample mean<br>

The standard deviation of a sampling distribution of a statistic is called the standard error of that statistic.<br>
Since we are dealing with the mean right now, that standard error of statistic is called the standard error of the sample mean,<br>
Standard error of the sample mean = sigma / sqrt(n) <br>
Sigma = true standard deviation of the population of all cartwheel distances <br>
sqrt(n) = factor of the sample size <br>
So the variability of our sample mean depends on these two features: 
1. how much variability was in the original population of measurements that we took our data from
2. And how much data did we take from that population <br>

Thus, if we were to have a larger sample size, the standard error would be smaller, meaning, our sample mean would vary less from the true mean. <br>
The **drawback** of the expression is that we can **never calculate it** as we don't know sigma, the true standard deviation of the population of cartwheel distances of **all adults**. Unless we look at ALL observations of ALL adults, we can't calculate it.<br>
So we have to come up with an estimate for this measure of variability, i.e. the standard error. we will use what we have from the sample. we can calculate the standard deviation of the sample and use it as our best estimate and we will use **s** instead of **sigma** <br>
Standard error of sample mean = **sigma**/sqrt(n) <br>

Estimated Standard error of sample mean = **s**/sqrt(n) <br>

Best estimate +- MoE <br>

Best Estimate = Unbiased Point Estimate ( Our sample mean in this case) <br>
MoE = 'a few' Estimated Standard errors. <br>
'a few' = multiplier from appropriate distribution based on desired confidence level and sample design <br>
95% Confidence Level -> 0.05 Significance.

Sample mean +- 'a few' * Estimated standard error of sample mean <br>
xbar +- t* (s/sqqrt(n)) <br>

as our n gets larger, the t* value gets closer and closer to it's Z* counterpart. <br>

If this procedure were repeated over and over, each time taking 25 random samples from that data and caclulating a 95% confidence interval estimate, we would have a lot of confidence intervals each of them with a 95% confidence interval behind them.
