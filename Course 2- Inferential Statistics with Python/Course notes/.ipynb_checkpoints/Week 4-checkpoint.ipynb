# Assumptions Consistency
**One Population Proportion**


Sample can be considered a simple random sample

Large enough sample size ()

* Confidence Interval: At least 10 of each outcome ()
* Hypothesis Test: At least 10 of each outcome ()


**Two Population Proportions**


Samples can be considered two simple random samples

Samples can be considered independent of one another

Large enough sample sizes ()

* Confidence Interval: At least 10 of each outcome ()
* Hypothesis Test: At least 10 of each outcome () - Where (the common population proportion estimate)


**One Population Mean**
Sample can be considered a simple random sample

Sample comes from a normally distributed population

* This assumption is less critical with a large enough sample size (application of the C.L.T.)


**One Population Mean Difference**
Sample of differences can be considered a simple random sample

Sample of differences comes from a normally distributed population of differences

* This assumption is less critical with a large enough sample size (application of the C.L.T.)


**Two Population Means**
Samples can be considered a simple random samples

Samples can be considered independent of one another

Samples each come from normally distributed populations

* This assumption is less critical with a large enough sample size (application of the C.L.T.)


Populations have equal variances – pooled procedure used

*  If this assumption cannot be made, unpooled procedure used

# Descriptive Inference examples for single variables using confidence intervals

## Example 1: Confidence intervals for proportions

### Research Question:
**What *Proportion* of non-Hispanic african americans age-18+ in the US in 2015-16 had systolic blood pressure greater than 130 mmHg? (Hypertension)**

### Inference Approach:
**Provide a 95% confidence interval estimate for this population proportion**

#### Step 1: Estimate the Population Proportion
* **Number** of black respondents with **non-missing data** on first systolic blood pressure measurement: 1135
* **Best points estimate:** Assuming simple random sample (srs) of black adults, Sample proportion is (465/1163) = 0.4097
    * 1163 : Total black adults
    * 465 : With systolic BP > 130
* **Interpretation:** Estimate 40.97% of all such black adults in 2015-16 had systolic blood pressure greater than 130 mmHg

#### Step 2:
**Computing the estimated standard error of our population estimate**
![step 2](img/step2-single-var-ci.png)

#### Step 3:
**Calculate the Confidence Interval**
![step 3](img/step3-single-var-ci.png)

##### Inference about the population proportion:
95% confidence interval for the population proportion of non-hispanic African anericans aged 18+ in us 2015-16 with systolic bp greater than 130 mmHg is: **(0.3811, 0.4384)**

**95% confidence** -> 95% of intervals formed this way are expected to cover the true population proportion!

**Inferenec:** if the hypothessized proportion was 0.35, 95% confidence interval suggests 0.35 is not a plausable value as it does not fall inside the confidence interval. This means we have evidence against the hypothesized proportion.

## Example 2: Condifence Interval for a mean

### Research Question:
**What was the mean Systolic blood pressure for non-hispanic african-americans age 18+ in us in 2015-16?

### Inference approach:
**Provide a 95% confidence interval estimate for this population mean**

#### Step 1: Estimate the population mean
* **Number** of black respondents with **Non-missing data** on the first systolic blood pressure measurement: 1135
* **Best point estimate:** Assuming srs of black adults, sample mean is 128.252 mmHg
* **Interpretation:** Our estimate of the mean systolic blood pressure for all such black adults in 2015-16 is 128.252 mmHg

#### Step 2: Calculate estimated standard error
**Standard error = sqrt(sampling variance of sample mean) = standard dev of the sampling distribution of all possible sample means if repeated samples of size 135 were obtained

**sample standard deviation** of the 1135 blood pressure measurements is 19.958 mmHg

Estimated standard error = 19.958 / sqrt(1135) = 0.592mmHg

#### Step 3: Form the confidence interval

**Best estimate +- MoE <br>
Best Estimate +- "A few" * Estimated SE <br>
for large sample size -> 95% multiplier = 1.96**

lower limit = 128.252 - 1.96 * 0.592 = 127.901 mmHg <br>
upper limit = 128.252 + 1.96 * 0.592 = 129.413 mmHg 

##### Inference about the population mean
95% confidence interval for the population mean systolic blood pressure of non Hispanic african americans age 18+ in us in 2015-16 is : **(127.091 mmHg, 129.413 mmHg)**

**Inference:** If hypothesized mean was 128 mmHg, 95% confidence interval suggests 128 is **a plausable value**

# Discriptive Inference Examples for single variables using Hypothesis Testing

## Example 1: One Sample Tests for Proportions

### Research Question:
**Did 33% (one-third) of non-Hispanic African-Americans age 18+ in the US in 2015-16 have systolic blood pressure greater than 130 mmHg, or was population proportion different than one-third?**
The arrangement of the question shows we want to test if the given value, 33% here, is correct or not.

### Inference Approach:
**Perform a one-sample test (two-tailed as it did not mention greater or less) to either reject or fail to reject the null hypothesis: Population proportion = 0.33**

#### Step 1: Define the Null and Alternative 
* Null: Population proportion (p) is 0.33
* Alternative: Population proportion is **Not equal** to 0.33

Alternative allows proportions to be either greater than or less than 0.33. This is a 2-tailed test and so needs more evidence against Null hypothesis to reject it. <br>
Significance level = alpha = 5% <br>
The 5% significance level means that the probability of rejecting the Null when it is true is 5%.

#### Step 2: Compute test statistic 
* **Best point estimate:** Assuming srs of black adults, sample proportion is 4.65 / 1135 = 0.4097
* **Test statistic:** Assuming sampling distribution of estimated proportion is normal, 
    * z = (best estimate - null value) / (null standard error) = (0.4097-0.33) / sqrt( (0.33*(1-0.33))/1163 )
    * Z = 5.71

**Z = 5.71** means that the Estimate is **More than** 5 standard errors **from 0.33, our null value.**

#### Step 3: Determind P-Value

![determine p-val](img/step-3-determind-p-val.png)

#### Step 4: Make a decision about the Null

* If population proportion really was 0.33 then observing a sample proportion of .4097 or more extreme is very unlikely.
* Since our p-value is much less than 0.05 significance level, strong evidence is present against Null -> We reject the null hypothesis.

**Bases on the estimated proportion of 0.4097, we support the population proportion of systolic blood pressure of black people in the US in the year 2015-16 is not 0.33 but most likely larger, much larger.

## Example 2: One sample Test for Means

### Research Question
**Was the mean sustolic blood pressure for non-hispanic african americans age 18+ in IS in 2015-16 equal to 128 mmHg or was the population mean -different- from 128?** <br>
Different -> Not equal to -> 2 tailed test

### Inference approach
**Perform a one sample test (2-tailed) to either reject or fail to reject null hypothesis: population mean = 128mmHg**

#### Step 1: Define null and alternative
* Null: Population mean (mu) is 128 mmHg
* Alternative: Population mean is not equal to 128 mmHg

Alternative allows proportions to be either greater than or less than 128. This is a 2-tailed test and so needs more evidence against Null hypothesis to reject it. <br>
Significance level = alpha = 5% <br>
The 5% significance level means that the probability of rejecting the Null when it is true is 5%. <br>
**As we are the ones conducting the study, we can choose alpha to be whatever we want.**

#### Step 2: Compute the Test Statistic

* **Best point Estimate:** Assuming srs of black adults, sample mean is 128.252 mmHg
* **Test Statistic:** Assuming sampling distribution of estimated mean is normal, 
    * t = best estimate - null value / estimated SE 
    * = (128.252 - 128) / (19.958 / sqrt(1135)) = 0.425
    
**Estimate is *less than* 1/2 standard errors from *128 null value* **

#### Step 3: Determine p-value
* If null hypothesis was true, would a test statistic value of only t = 0.425 be unusual enough to reject the null?
* p-value - probability of seeing test statistic of 0.425 or more extreme assuming the null hypothesis is true/

![step 3](img/step-3-p-val-one-sample-test-for-means.png)

#### Step 4: Make a decision about the Null
* If population mean really was 128 mmHg, then observing a sample mean of 128.252 or more extreme is quite likely.
* Since our p-value is much bigger than 0.05 significance level, weak ecidence against the null -> we **Fail to reject the Null**

**We don't accept null but fail to reject it**

Based on estimated mean of 128.252 mmHg, we **Cannot support** the claim that the population mean differs from 128 mmHg.

# Comparing means for two independent samples

**Now, we will turn to comparing different (2 independent) sub groups in a population instead of estimating one overall population. <br>
We will focus on comparing means of two independent subgroups in the population**

## Example: Comparing means in two groups

### Research question
**Considering African american adults Living in the US in 2015-16, did males and females have significantly different mean systolic blood pressure?**

### Inference Approaches:
* Form a confidence interval for the difference in the two means
* Perform a two sample t-test for difference in the two means
* Check assumptions

#### Approach 1: Form a confidence Interval
**Males:** Mean = 131.01, std dev = 20.59, n=536 <br>
**Females:** Mean = 125.79, standard dev = 19.06, n = 599 <br>

* **Best point estimate:** Difference in sample means is 131.01 - 125.79 = 5.22 mmHg 
* **Interpretation:** In 2015-16 we estimate the mean systonic blood pressure for all emale black adults was 5.22 mmHg **Higher** than that for all female black adults.
Yes, it's a higher mean, but we want to know if that is a **Significant** Higher mean.

**Note:** Sample standard deviations are similar. <br>
* We will now **check some assumptions** and decide which **confidence interval approach** is reasonable, **pooled** or **unpooled**.
![Checkign assumptions](img/check-assumption-approach-1.png)

<br><br>

![Checkign assumptions](img/check-assumption-2-approach-1.png)

##### Forming a confidence interval on approach 1:
* Assuming equal variance (pooling):
    * 95% Ci for difference in means = (2.91 mmHg, 7.53 mmHg)
    * Interval doesn't include **zero (0)**, -> **Significant Difference**
* Assuming uneqqual variance (No pooling): 
    * 95% CI for difference in means = (2.90, 7.54)
* Same Conclusion! 
    * Results are robust to possible violation of this assumption

**So with 95% confidence, we can say that black males had a higher systolic bp compared to black females in the US in 2015-2016**

#### Approach 2: Two sample t-test
* **Null:* Males and Females have equal population means
* **Alternative:** Males and females have **Different** means
   
- Alternate allows male mean to be either greater or less than female mean.
    - Two tailed test needed for more evidence against null hypothesis to reject it

* Significance level = 5%

##### Assumptions:
* **Normal** distribution of blood pressure in each population
    * may not be reasonable, based on previous histograms and QQ plots for each sample
* **Same Standard Deviation** in each population
    * Somewhat reasonable and these two sampling techniques can be robust to violations of this assumption but we can examine both pooled and unpooled test results
    
**Assuming equal variance (pooling):**
    * T = 4.436, df = 1133 (1135 - 2 degrees of freedom), p-value < 0.001
        * We **Reject the Null hypothesis**, the means are different
**Assuming unequal variance (unpooled):**
    * t = 4.417, df = 1097.43, p-value < 0.001

**Same conclusion! Our results are ribust to possible violation of assumption**

###### What if normality doesn't hold?
* Not convinced the variable of interest follows a normal distribution in each population?
    * **Non parametric test** that does not assume normality
* Non-parametric analog of two sample t-test
    * = **Mann-Whitney test**
    * Compares location of distributions using medians
    
    * **nn-Whitney test result: p-value < 0.001**
        * We reject the Null Hypothesis that both distributions have identical locations
    * COnclusion is robust to potential violations of normality


#### Consistent evidence of robust difference in central tendencies of the two distributions, regardless of assumptions made and approach to inference used
