# Where data come

### Types
1. Organic / Process data
    1.1 Data that are generated by a computer information system or are recorded by video or audio. They are generated organically and are often generated over time.
    1.2 Examples:
        1.2.1 Financial or point of sale transactions/Stock Market exchanges
        1.2.2 Netflix viewing history
        1.2.3 web browser activity
        1.2.4 temperature/pollution sensors
    1.3 These prpcesses generate massive quantities of data : Big Data
    
2. "Designed" Data Collection
     2.1 Designed to specifically address a stated research objective
         2.1.1 Individuals sampled from a population, interviewed about opinions on a particular topic
         2.1.1 twitter tweet collection analysis :- focus on a specific trend
     2.2 Common features of designed data
         2.2.1 Sampling from populations, administration of carecully designed questions
         2.2.1 typically data sets much smaller compared to organic/process data sets
         2.2.3 data collected for very specific reasons, rather than simple reflections of ongoing natural processes


## Can the data be considered i.i.d.?
i = independend and id = identically distributed

independent - It means that each observation is independent of all other observations
identically distributed - Each of the observation is measured using a common statistical distribution

e.g:
Final exam scores from a large intro to stats class at a university are independent observations from a common normal distribution, i.e. each student is treated independently and all measure th ecommon bell shaped curved (normal distribution)
<br><br>
If we assume a data is iid, it helps us perform certain statistical measures to a certain degree of precission

## What if it is not iid


e.g: students in class may cheat :- have similar class
males and females have different means, thus males and females have different distributions
students from same discussion sections may have similar scores

Dependencies and differences need to be accounted for in analysis!
Need different analytic procedures

## Need to ask:
Can we apply procedures that assume i.i.d data?

# Types:
1. Quantitive Variables
    1.1 Numerical, measurable quantitis in which arithmetic operations often make sense
    1.2 Types: 2
        1.2.1 Continuous - could take on any value within an interval, many possible values : age, time it took etc
        1.2.2 Discrete - countable value, finite number of values : No. of childten in a household
2. Qualitative Variables
    2.1 classifies individuals or items into different groups
    2.2 Types: 2
        2.2.1 Ordinal: Has some sort of order or ranking associated with it. E.g. Class ranking in highschool or collage \
        (Freshman/sophomores, juniors/seniors)
        2.2.2 Nominal: Groups are merely names, no ranking. E.g. Races, Marital status, native country in numbers etc. 
        As we move around the different categories in the variable, it does not change the meaning.

Types of Research Studies:
1. Exploratory versus Confirmatory studies <br>
    1.1 Confirmatiory: Scientific method ~ specify falsifiable hypothesis, then test it -> collect data to address single pre-specified qquestion
    1.2 Collect and analyse data without first pre-specifying question
        1.2.1 CAUTION: Can be informative but **watch out** for **"overfitting", "multiple testing", "p-hacking"**
        1.2.2 The more questions you ask from a dataset -> the more likely you are to draw a misleading conclusion.
    
2. Comparative versus Non-Comparative studies
    2.1 Comparative: **Goal = Contrast** one quantity to another. 
        <br> Compare yealds of oranges in plots of land with different fertilizers.
        <br> Preference of voters for one candidate for the office from another.
        <br> Compare the rate at which viewers would click on one of 2 different versions of an online ad.
    2.2 Non- Comparative: **Focul = estimating or predicting** absolute quantities ~ not (explicitely) comparative
        <br> e.g.: 
        <br> predict the value of the company's stock one year from now
        <br> Estimate the reduction of blood pressure followed by treatment by a certain drug
3. Observational studies versus Experiments
    3.1 Observational: arises **naturally**, contrasts based on **self selection** of units into groups
    3.2 Experimental: involve **manipulation or assignment** -> experimenter deliberately treats different units in different
        ways
